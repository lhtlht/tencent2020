```python
#from utils import *
import pandas as pd
import numpy as np

path_buld = "../../data/tencent2020/build/"
path_save = "../../data/tencent2020/save/"


train_click_log = pd.read_csv(path_save + "train_click_log.csv", encoding="utf-8")
test_click_log = pd.read_csv(path_save + "test_click_log.csv", encoding="utf-8")
```

```python
print(train_click_log.shape)
train_click_log.head()
```

```python
train_click_log.sort_values(by=['user_id','time'], ascending=[True,True], inplace=True)
train_click_log = train_click_log.reset_index(drop=True)
train_click_log.head()
```

```python
def to_list(ll):
    return ','.join(ll.astype(str))

train = train_click_log.groupby('user_id').agg({'creative_id':to_list,
                                                'ad_id':to_list,
                                                'product_id':to_list,
                                                'click_times':to_list,
                                                'time':to_list,
                                                'product_category':to_list,
                                                'advertiser_id':to_list,
                                                'industry':to_list}).reset_index()
test = test_click_log.groupby('user_id').agg({'creative_id':to_list,
                                                'ad_id':to_list,
                                                'product_id':to_list,
                                                'click_times':to_list,
                                                'time':to_list,
                                                'product_category':to_list,
                                                'advertiser_id':to_list,
                                                'industry':to_list}).reset_index()
train.to_csv('train.csv', index=False, encoding='utf-8')
test.to_csv('test.csv', index=False, encoding='utf-8')
```
